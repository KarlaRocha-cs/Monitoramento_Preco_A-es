```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ticker = 'AAPL' #Exemplo: Apple Inc.
#Obtenha os dados históricos da ação usando o yfinance
data = yf.download(ticker, start='2022-01-01', end='2022-12-31')

# 2. Processamento de dados com Pandas e Numpy:
# Aqui, como o yfinance retorna automaticamente um DataFrame, não precisamos fazer muitas transformações.
# Se necessário, você pode realizar manipulações adicionais usando as funções do Pandas e NumPy.

# 3. Visualização de dados com Matplotlib:
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Preço de fechamento') #Plotagem do preço de fechamento
plt.legend()
plt.title('Histórico de Preços de ' + ticker)
plt.xlabel('Data')
plt.ylabel('Preço de fechamento')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import yfinance as yf

# Baixar dados da AAPL como exemplo
dados_aapl = yf.download('AAPL', start='2020-01-01', end='2023-01-01')

# Calcular o retorno diário
dados_aapl['Retorno Diário'] = dados_aapl['Adj Close'].pct_change()

# Calcular o retorno cumulativo
dados_aapl['Retorno Cumulativo'] = (1 + dados_aapl['Retorno Diário']).cumprod()

print(dados_aapl[['Adj Close', 'Retorno Cumulativo']].head())
```

```python
# A. Importe as bibliotecas
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# B. Defina os tickers das ações (GOOGLE E MICROSOFT)
tickers = ['GOOGL', 'MSFT']

# C. Obtenha os dados históricos das ações usando o yfinance
data = yf.download(tickers, start='2020-01-01', end='2023-01-01')['Adj Close']

# D. Calcule o retorno diário para cada ação
retorno_diario = data.pct_change()

# E. Calcule o retorno cumulativo para cada ação
retorno_cumulativo = (1 + retorno_diario).cumprod()

# F. Calcule o retorno médio diário para cada ação
retorno_medio_diario = retorno_diario.mean()

# G. Calcule o desvio padrão dos retornos diários para cada ação
desvio_padrao_diario = retorno_diario.std()

# H. Imprima os resultados
print("Retorno Médio Diário:")
print(retorno_medio_diario)
print("\nDesvio Padrão dos Retornos Diários:")
print(desvio_padrao_diario)

# I. Crie um gráfico de linha para visualizar os retornos cumulativos das ações
plt.figure(figsize=(12, 6))
plt.plot(retorno_cumulativo, label=retorno_cumulativo.columns)
plt.title('Retorno Cumulativo das Ações')
plt.xlabel('Data')
plt.ylabel('Retorno Cumulativo')
plt.legend()
plt.grid(True)
plt.show()
```
